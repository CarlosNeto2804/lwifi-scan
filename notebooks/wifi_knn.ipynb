```python

# Obter o dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix

df = pd.DataFrame(pd.read_pickle('../wifi_dict.pkl'))
X = df[df.columns.difference(['COMODO'])].values.astype(int)
Y = df['COMODO'].values 
```

```python
def show_results(n_neighbors, s_accuracy, s_precision, s_recall, f1, conf_matrix):
    print("n_neighbors: {ngh}".format(ngh=n_neighbors))
    print("precision_score {pre}: ".format(pre=s_precision))
    print("recall_score: {re} ".format(re=s_recall))
    print("f1_score: {f1}".format(f1=f1) )
    print("accuracy_score {acc}".format(acc=s_accuracy))
    print("confusion_matrix:\n {conf}".format(conf=conf_matrix))
```

```python
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.3, train_size=0.7,random_state=9)
best_neighbors = {}
for i in range(1,13):
    knn = KNeighborsClassifier(n_neighbors = i)
    x_normalized = (X - X.mean(axis=0)) / X.std(axis=0)
    accuracy_list = []

    for _, test_index in sss.split(x_normalized,Y):
        x_test, x_validation, y_test, y_validation = train_test_split(x_normalized[test_index],
                                                                                    Y[test_index],
                                                                                    test_size=0.5,
                                                                                    random_state=0)

        knn.fit(x_validation, y_validation)
        y_predicted = knn.predict(x_test)
        __, std, mean, s_f1 = get_metrics(i, accuracy_list, y_test, y_predicted)
        best_neighbors[s_f1] = (i, std, mean)

best_neighbor = best_neighbors[max(best_neighbors)]
```
